```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('Motorcycledata.txt')
t=data[:,1]
y=data[:,2]
t_=np.array(sorted(list(set(t))))
m=list([])
y_=[]
for ti in t_:
    y_.append(np.mean(y[t==round(ti,1)]))
    m.append(len(y[t==round(ti,1)]))
print(len(m))
print(len(y_))
print(len(t_))

for k in range(3,7):
    coeffs=np.polyfit(t_,y_,k)
    y_e=np.poly1d(coeffs)(t_)

    plt.figure(1)
    plt.scatter(t_,y_)
    plt.plot(t_,y_e,'--')
plt.grid()
plt.show()
```

```python
# regression spline
n=len(t_)
h=np.zeros(n-1)
Q=np.zeros([n,n-2])
R=np.zeros([n-2,n-2])
for i in range(n-1):
    h[i]=t_[i+1]-t_[i]
for j in range(1,n-1):
    Q[j-1,j-1]=1/h[j-1]
    Q[j,j-1]=-1/h[j-1]-1/h[j]
    Q[j+1,j-1]=1/h[j]
for i in range(1,n-1):
    R[i-1,i-1]=1/3*(h[i-1]-h[i])
for i in range(1,n-2):
    R[i-1,i]=1/6*h[i]
    R[i,i-1]=1/6*h[i]

K=np.dot(np.dot(Q,np.linalg.inv(R)) , np.transpose(Q))
for alpha in [10,20,30]:
    g=np.dot( np.linalg.inv((np.eye(len(K)) + alpha*K)) , y_)
    plt.figure(2)
    plt.scatter(t_,y_)
    plt.plot(t_,g,'--')
plt.grid()
plt.show()
```

```python
for alpha in range(1,50):
    g=np.dot( np.linalg.inv((np.eye(len(K)) + alpha*K)) , y_)
    w=1/len(y)*np.array(m)
    W=np.diag(w)
    Aw=np.dot(np.linalg.inv(W+alpha*K) , W)
    CV=np.sum(w*((y_-g)/np.diag(np.eye(n) - Aw))**2)
    plt.figure(3)
#     plt.scatter(t_,y_)
    plt.scatter(alpha,CV)
plt.grid()
plt.show()
```
